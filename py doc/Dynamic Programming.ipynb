### Fibonacci Problem

```python
def fibonacci(n):
    if n < 0:
        return -1
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
    
start = time.time()
res = fibonacci(40)

print("Execution Time: %s Seconds" %(time.time() - start))
print(res)
```

```python
def fibonacci_2(n):
    if n < 0:
        return -1
    a, b = 0, 1
    for i in range(n):
        a,b = b, a+b
    return a
    
start = time.time()
res = fibonacci_2(40)

print("Execution Time: %s Seconds" %(time.time() - start))
print(res)
```

```python
# Fibonacci problem using general recursion , exponential time complexity O(2^n)
def fib_recursion(n):
    #start = time.time()
    if n == 1 or n==2:
        result = 1
    else:
        result = fib_recursion(n-1) + fib_recursion(n-2)
    #end = time.time()
    #e_time = end - start
    return result

start = time.time()
res = fib_recursion(40)

print("Execution Time: %s Seconds" %(time.time() - start))
print(res)
```

```python
# Fibonacci with Memoized solution, linear time O(2n) = O(n)
def fib_memo(n,memo):
    if memo[n] != None:
        return memo[n]
    if n== 1 or n==2:
        result = 1
    else:
        result = fib_memo(n-1,memo) + fib_memo(n-2,memo)
    memo[n] = result
    return result


start = time.time()
res = fib_memo(40,[None] * 41)

print("Execution Time: %s Seconds" %(time.time() - start))
print(res)
```

```python
def next_ugly(n):
    ugly = [0] * n
    ugly[0] = 1
    i2 = i3 = i5 = 0
    
    next_2, next_3, next_5 = 2, 3, 5
    
    for i in range(1,n):

        ugly[i] = min(next_2, next_3, next_5)
        
        
        if ugly[i] == next_2:
            i2 += 1
            next_2 = ugly[i2] * 2
            
        if ugly[i] == next_3:
            i3 += 1
            next_3 = ugly[i3] * 3
        
        if ugly[i] == next_5:
            i5 += 1
            next_5 = ugly[i5] * 5
    return ugly[-1]
    
    
    
    
    
print(next_ugly(150))
```

```python
# Fibonacci Problem, space and time complexity O(n)

def fib(n, cache={}):
    if n <= 1:
        return n
    elif n not in cache:
        cache[n] = fib(n-1) + fib(n-2)
    return cache[n]

fib(5)
```

```python
# Fibonacci, space (1)
def fib_iterative(n):
    if n <= 1:
        return n
    f_2, f_1 = 0, 1
    for _ in range(1,n):
        f = f_2 + f_1
        f_2, f_1 = f_1, f
    return f_1

fib_iterative(5)
```

```python
# Find maximum subarray in an array

import itertools

def find_max_subarray(A):
    min_sum = max_sum = 0
    for run_sum in itertools.accumulate(A):
        print(run_sum)
        min_sum = min(min_sum, run_sum) # for negative numbers
        print(min_sum)
        max_sum = max(max_sum, run_sum - min_sum)
        print(max_sum)
        print('--------')
    return max_sum
```
